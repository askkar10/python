what is a module

```python
#a modeule is a file containing code
%%writefile mymath.py
"""mymath - our example math module"""
def _b(x):
  return x
pi = 3.14159
def area(r):
    """area(r): return the area of a circle with radius r."""
    return(pi * r * r)
```

```python
import mymath
print(mymath.pi)
print(mymath.area(10))
```

```python
from mymath import pi,area
print(pi)
print(area(12))
```

private names in modules

```python
#a modeule is a file containing code
%%writefile private.py
"""mymath - our example math module"""
_b = 3
def _c(x):
  return x
```

```python
import private
print(private._b)
print(private._c(10))
```

```python
from private import _b,_c
print(_b)
print(_c(100))
```
