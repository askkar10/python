handling sctructured binary data with the struct module

```python
# struct module -> rading and writing binary data - tutaj aktualnie zapisanie
import struct record_format = 'hd75'
data_record = struct.pack(record_format,42,3.14,b'goobye')
data_file = open('data','wb')
data_file.write(data_record)
# h -> single c short integet
# d -> single c double precision floating point
# s -> string
# 7s -> string zawierajacy 7 znaków
```

```python
# otwieranie takiego pliku
import struct
record_format = 'hd7s'
record_size = struct.calcsize(record_format) # calcsize to otwieranie takich plików
result_list = []
with open('data','rb') as input:
  while True:
    record = input.read(record_size)
    if not record:
      break
    result_list.append(struct.unpack(record_format,record))
print(result_list)
```

pickling objects to files

```python
# pickle -> pozwala na przekształcenie struktury obiektu Pythona (listy itp.) w strumień bajtów, który mozna
# następnie zapisać do pliku, przesłać przez sieć do bazy danych itp.

# unpicling to operacja odwrotna

# saving data pickle.dump zapisuje wszystko

import pickle
a = 42
b = 3.14
c = 'test'
file = open('state','wb')
pickle.dump(a,file)
pickle.dump(b,file)
pickle.dump(c,file)
file.close()
```

```python
# odzyskiwanie tych danych

import pickle
file = open('state','rb')
a = pickle.load(file)
b = pickle.load(file)
c = pickle.load(file)
file.close()
print(f"{a=} {b=} {c=}")
```

```python
# zapisywanie i otwieranie za pomocą funkcji
import pickle
def save_data(data_dict):
  with open('state2','wb') as file:
    pickle.dump(data_dict,file)

def restore_data():
  with open('state2','rb') as file:
    data_dict = pickle.load(file)
  return data_dict

if __name__ == '__main__':
  data_dict = {'a':42,
               'b':3.14,
               'c':'test'}

save_data(data_dict)
restored_data = restore_data()
print(restored_data)
```
