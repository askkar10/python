defining new exceptions

```python
class MyError(Exception):
  pass

# raise MyError('some information about what went wrong')

try:
  raise MyError('some information about what went wrong')
except MyError as error:
  print('situation', error)
```

```python
# multiple error argument
class MyError(Exception):
  pass

try:
  raise MyError('some information','my_filename',3)
except MyError as error:
  print('Situation: {0} with file {1}\nerror code: {2}'.format(
      error.args[0],error.args[1],error.args[2]
  ))
```

exceptions groups

```python
# exception group => multiple exceptiion by by except*
message = ''
try:
  raise ExceptionGroup("Multiple exceptions",[TypeError(),FileNotFoundError(),ValueError()])
except* TypeError:
  message += f"Handling TypeError\n"
except* IOError:
  message += f"Handling IOError\n"
except* ValueError:
  message += f"Handling ValueError"
finally:
  print(message)
```

context managers using th with keyword

```python
# normalnie byśmy musieli zamknąć plik i byłby długi kodzik
try:
  infile = open(filename)
  data = infile.read()
finally:
  infile.close()
```

```python
# mozemy to zrobić zdecydowanie szybciej
with open(filename) as infile:
  data = infile.read()

print(data)
```

```python
# tak mozemy skopiowac
with open('empty.txt') as infile, open('other.txt','w') as outfile:
  data = infile.read()
  outfile.write(data)

# lub

with (
    open('empty.txt') as infile,
    open('other.txt') as outfile
):
  data = infile.read()
  outfile.write(data)
```

example: exceptions in normal evaluation

```python
# jezeli damy normalny cyfrą to zwróci ją jako float, jeżeli damy pusty string "" to bedzie 0, a jezeli uzupelniony to zwróci None
def cell_value(string):
    try:
        return float(string)
    except ValueError:
        if string == "":
            return 0
        else:
            return None
print(cell_value(""))
```

```python
def safe_apply(function,x,y):
  try:
    return function(x,y)
  except TypeError:
    return None
def dodawanie(x,y):
  return x + y
print(safe_apply(dodawanie,12,40))
```
