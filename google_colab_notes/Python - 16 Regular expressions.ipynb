extracting method text from strings

```python
# surname, firstname middlename: phonenumber
# [-a-zA-Z] -> single hyphen, single letter, or a single uppercase ltter
# [-a-zA-Z]+ -> matching full name such as McDonald
# \d\d\d-\d\d\d-\d\d\d\d -> \d -> matching any digit -> this matchin phone number like 123-123-1234
# (\d\d\d-)?\d\d\d-\d\d\d\d -> (\d\d\d)? is optional -> rest normak
# (\d{3}-)?\d{3}-d{4} -> tak jak wyzej tylko ze skrocone d{4} oznacza cztery cyfry
import re
regexp = re.compile(r'[-a-zA-Z]+,'
                    r' [-a-zA-Z]+'
                    r'( [-a-zA-Z]+)?'
                    r': (\d{3}-)?\d{3}-d{4}'
                    )
file = open("textfile3",'r')
for line in file.readlines():
  if regexp.search(line):
    print('Yeah i found so what')
    break
  file.close()
```

```python
# ?P<name> grupowanie regexpow
# result.group()
import re
regexp = re.compile(r'(?P<last>[-a-zA-Z]+,)'
                    r' (?P<first>[a-zA-Z]+)'
                    r'( ?P<middle>[a-zA-Z]+)'
                    r': (?P<phone>(\d{3}-)?\d{3}-d{4})')
file = open('textfile3','r')
for line in file.readlines():
  result = regexp.search(line)
  if result == None:
    print(r"Ops i don't this this is a rcord")
  else:
    last_name = result.group('last')
    first_name = result.group('first')
    middle_name = result.grop('middle')
    if middle_name == None:
      middle_name = ""
    phone_number = result('phone')
    print(f"Name: {first_name} middle name: {middle_name} last name: {last_name} number: {phone_number}")
file.close()
```

using a function with sub

```python
import re
int_string = "1 2 3 4 5"
def init_match_to_float(match_obj):
  return(match_obj.group('num') + ".0")
pattern = r"(?P<num>[0-9]+)"
regexp = re.compile(pattern)
regexp.sub(init_match_to_float, int_string)
```
