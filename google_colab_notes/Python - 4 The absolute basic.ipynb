complex numbers

```python
# nj -> n to normalna liczba
print(3+2j - (4+4j))
print((1+2j) * (3+4j))
z = (3+5j)
print(z.real) # -> real part
print(z.imag) # -> imagine part
```

```python
import cmath
# https://docs.python.org/3/library/cmath.html
print(cmath.sqrt(complex(-2.0, -0.0)))
print(cmath.phase(complex(-1.0, -0.0)))
```

*   class names -> Capitalize Each Word -> MyClass


*   constants values -> ALL_CAPS_WITH_UNDERSCORES -> PI, TAX_RATE
