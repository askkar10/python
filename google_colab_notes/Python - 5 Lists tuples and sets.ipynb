list indices

```python
x = ["first","second","third","fourth"]
print(x[0])
print(x[2])
print('-------')
print(x[-1])
print(x[-3])
print('-------')
print(x[1:-1])
print(x[0:3]) # lub print(x[:3])
print(x[-2:-1])
print(x[-1:2])
print('-------')
print(x[:2])
print('-------')
# kopiowanie listy
z = x[:]
z[0] = '1st'
print(z)
```

modifying list

```python
x = [1,2,3,4]
x[1] = "two" # zmiana drugiego elementu listy
print(x)

print('-------')
x = [1,2,3,4]
x[len(x):] = [5,6,7]
print(x)
x[:0] = [1,0]
print(x)
print('-------')
x[1:-1] = []
print(x)

print('-------')
x = [1,2,3]
x.append("four") # dodanie na koniec 'four'
print(x)

print('-------')
x = [1,2,3,4]
y = [5,6,7]
x.extend(y) # rozszerzenie listy x o elementy listy y
print(x)

print('-------')
x = [1,2,3]
x.insert(2,"Hello") # dodanie 'hello' na trzecia pozycje w liscie
print(x)

x.insert(0,"start")
print(x)

print('-------')
x = ['a',2,'c',7,9,11]
del x[1] # usnięcie drugiego elementu
print(x)
del x[:2] # usnięcie pierwszego drugiego elementu
print(x)

print('-------')
x = [1,2,3,4,3,5]
x. remove(3) # usunięcie pierwszego wystąpienia 3
print(x)
x.remove(3)
print(x) # usniecie drugiego wystapienia 3

print('-------')
x = [1,3,5,6,7]
x.reverse()
print(x)
```

sorting list

```python
x = [3,8,4,0,2,1]
x.sort() # To zmienia listę, która sortujesz, żeby jej nie zmienić trzeba skopiować lub użyć sorted()
print(x)
print("-----")
x = [2,4,1,3]
y = x[:]
y.sort()
print(x)
print(y)
print("-----")
y.sort(reverse=True)
print(y)
```

custom sorting

```python
def comp_num_of_chars(string1):
  return len(string1)
word_list = ['Python','is','better','than','C']
word_list.sort()
print(word_list)
print("-----")
word_list.sort(key=comp_num_of_chars)
print(word_list)

# użycie słowa kluczowego "key" + funckja
```

the sorted() function

```python
x = [4,3,2,1]
y = sorted(x) # Sortowanie listy o ona zostaja taka sama + słowo kluczowe key tez moze tutaj wystapic
print(y)
print("-----")
y = sorted(x,reverse=True)
print(y)
```

nested list and deep copies

```python
m = [[0,1,2],[10,11,12],[20,21,22]]
print(m[0])
print(m[0][1])
print(m[2])
print(m[2][2])
print("-----")
nested = [0]
original = [nested,1]
print(original)
print("-----")
nested = 'zero'
print(original)
print("-----")
original = [[0],1]
shallow = original[:]
import copy
deep = copy.deepcopy(original)

shallow[1] = 2
print(shallow)
print(original)
shallow[0][0] = 'zero' # przy zwykłym skopiowaniu zmieni wartość originalnej i kupiowanej
print(shallow)
print(original)
deep[0][0] = 5 # trzeba uzyc glebokiej kopi zeby to sie nie moglo skopiowac
print(deep)
print(original)
```

tuples

są podobne do listy, różnica jest taka, ze nie można edytować elementu tuple, może być tylko stworzony

tuple basic

```python
x = ('a','b','c')
print(x[2])
print(x[1:])
print(len(x))
print(max(x))
print(min(x))
print(5 in x)
print(5 not in x)
# x[2] = 'd' -> error nie można zmienić wartości

print(x + x)
print(2 * x)
```

tuple same w sobie nie moga byc modyfikowane, lecz gdy maja w sobie obiekty(listy dictionary) to wtedy mozna te obiekty edytowac

packing and unpacking tuples

```python
from os import XATTR_REPLACE
(one,two,three,four) = (1,2,3,4)
print(one,two,three,four)
one,two,three,four = 5,6,7,8
print(one,two,three,four)
# unpacking
x = (1,2,3,4)
a,b,*c = x
print(a,b,c)
a,*b,c = x
print(a,b,c)
*a,b,c = x
print(a,b,c)
a,b,c,d,*e = x
print(a,b,c,d,e)
# packing and unpacking can also be performed with list
[a,b] = [1,2]
[c,d] = [3,4]
[e,f] = (5,6)
(g,h) = 7,8
i,j = [9,10]
k,l = (11,12)
print(a)
print(b,c,d)
print(e,f,g)
print([h,i,j,k,l])
print((h,i,j,k,l))
```

sets

jest to nieuporządkowana kolekcja obiektów, uzywana dzięki unikalności, ponieważ w "set" elementy się nie powtarzają

```python
x = set([1,2,3,1,3,5])
print(x)
x.add(6)
print(x)
print(1 in x)
print(4 in x)

y = set([1,7,8,9])
print(x | y) # kombinacja dwóch setów
print(x & y) # inversion, czyli te które się powtarzja
print(x ^ y) # znajduje elementy ktore sie nei powtarzaą
```

frozen sets
