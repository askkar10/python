```python
print(" ".join(['join','puts','spaces','between','elements']))
print('----')
print("::".join(['separated','with','colons']))
print('----')
print("".join(['separated','by','nothing']))
print('----')
x = 'you\t\t can have tabs\t\t and new lines\n\n mixed'
print(x.split())
print('----')
x = 'mississipi'
print(x.split('ss')) # po danej wartosci
print('----')
x = 'a b c d'
print(x.split(' ',1)) # tylko jedną
print(x.split(' ',2)) # dwie
print(x.split(' ',9)) # dziewiec
```

getting rid of extra whitespace

```python
# strip - bez lewej i prawej spacji
# rstrip - bez prawej
# lstrip - bez lewej
x = '     hello,          world\t\t'
print(x.strip())
print(x.lstrip())
print(x.rstrip())

import string
string.whitespace # informuje o tym co w stringu jest biala spacja
# za  pomoca strip,lstrip,rstrip możemy usuwać znaki w zdaniu
print('----')
x = 'www.python.org'
print(x.lstrip('w'))
print(x.rstrip('gor'))
print(x.strip('.wrgo')) # nie musimy być po kolei
print('----')
# jezeli chcemy aby usuwalo poczatek albo koniec to: removepreffix lub remove suffix
print(x.removeprefix("www."))
print(x.removesuffix(".org"))
```

string searching

```python
# najpotezniejsze narzedzie to -> re -> zostanie omowione w pozniejszych rozdzialach
# find, rfind, index, rindex, count
x = 'mississipi'
print(x.find("ss")) # znajduje na jakim indexie się znajduje
print(x.find('zz')) # jezeli nie ma to -1
print(x.find('ss', 4))
print(x.find('ss',0,4)) # kolejne parametry to start i end od którego ma sie zaczac wyszukiwanie
print('----')
print(x.rfind('ss')) # wyszukuje od prawej do lewej
# index,rindex -> to samo tylko, jeżeli nie znajdzie to wystąpi error
print('----')
# count -> liczy ile razy wystapi dany parametr
print(x.count('ss'))
print('----')
# startswith,endswith -> sprawdza czy dane slowo sie zaczyna lub konczy tak jak podany paramatre
# mozemy tez podac kilka i bedzie to jako tuple
print(x.startswith('miss'))
print(x.endswith('ipi'))
print(x.endswith(('i','u')))
```

modifying strings

```python
# replace -> zamienia znaki na podany parametr
x = 'mississipi'
print(x.replace('ss','+++'))
print('----')
# string.maketrans, string.translate -> mozemy dzieki temu zmieniac dane wyrazy lub pojedyncze slowoa
x = '~x^(y%z)'
table = x.maketrans("~^()","!&[]")
print(x.translate(table))
print('----')
# mozemy tez usuwac
table = x.maketrans("","",'~^()%') # trzeba podac dwia puste parametry
print(x.translate(table))
print('----')
x = 'mISSissipi'
print(x.lower()) # male znaki
print(x.upper()) # duze znaki
print(x.title()) # pierwsza litera duze
print(x.swapcase()) # zamiana malych na duze i odwrotnie
```

the format method and positional parameters format

```python
print("{0} is the {1} of {2}.".format('Ambrosia','food','the gods'))
print("{{Ambrosia}} is the {0} of {1}".format('food','the goods'))
# jezeli chcemy aby nie zaznaczylo miejsca do zmiany musimy podać w {{}}
```

the format method and named parameters

```python
print("{food} is the food of {user}.".format(food="Ambrosia",user="the gods"))
print("{0} is the food of {user[1]}.".format('Ambrosia',user=['people','the gods','askaniusz']))
# mozemy podawac parametry ktore chcemy uzyc i przypisac ich do zmiennych
```

format specifies


```python
print("{0:10} is the food of the gods".format('ambrosia'))
print("{0:{1}} is the food of the gods".format('ambrosia',20))
print("{0:>30} is the food of the gods".format('ambrosia'))
print("{0:&>25} is the food of the gods".format('ambrosia'))
```

string interpolation with f-strings

```python
value = 42
message = f"The answer is {value}"
print(message)
print('----')
primes = [1,2,3,5]
print(f"sum of first 4 primes is {sum(primes)}")
print('----')
PI = 3.1415
print(f"PI is {PI:{10}.{2}}") # pierwsze {} ile spacji od ostatniego slowa, drugie {} to do ilu miejsc po precinku sformatowac
print('----')
print(f"sum of first 4 primes is {sum(primes)=}")
```

formatting strings with %

```python
print("%s is the %s of %s" % ('ambrosia','food','the goods'))
x = [1,2,3]
print("the %s contains %s" % ("list",x))
```

named parameters and formatting sequences

```python
num_dict = {'e': 2.718,'PI':3.14159}
print("%(PI).2f" % num_dict)
```
