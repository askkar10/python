other dictionary operations

```python
english_to_french = {}
english_to_french['red'] = 'rouge'
english_to_french['blue'] = 'bleu'
english_to_french['green'] = 'verte'
print(len(english_to_french)) # wypisane dlugosci dictionary
key_list = list(english_to_french.keys()) # wypisanie kluczy do listy
print(key_list)
value_list = list(english_to_french.values()) # wypisanie wartosci do listy
print(value_list)
key_value_list = list(english_to_french.items()) # wypisanie kluczy,wartosci do listy
print(key_value_list)
print('red' in english_to_french) # sprawdzenie czy dana wartosc wystepuje lub nie w liscie
print('red' not in english_to_french)
print(english_to_french.get('red','No translation')) # sprawdzenie czy dany klucz istnieje, jezeli tak to wypisuje wartosc, jezeli nie to zwraca drugi parametr
print(english_to_french.get('orange','No translation'))
print(english_to_french.setdefault('orange','No translation')) # sprawdza czy dany klucz istnieje jezeli nie ma to, tworzy nowy klucz i przypisuje wartosc drugiego parametru
print(english_to_french) # mozna uzyc tez 'defaultdic' po zaimportowaniu 'collections' - dziala tak samo

y = {0:'zero',1:'one'}
x = y.copy() # tak mozemy skopiowac dictionary, mozemy tez uzyc deepcopy
del x[0] # usuniecie klucza
print(y,x)
x = {0:'jeden',1:'two',4:'gucci'}
y.update(x) # aktualizuje liste jezeli wystepuja te same wartosci to po prostu zamienia jej value, a jezeli jest nowy klucz to dodaje key:value
print(y)
```

word counting

```python
# liczenie ile razy dane slowo wystepuje w stringu
sample_string = 'to be or not to be'
occurrences = {} # stworzenie pustego dictionary
for word in sample_string.split(): # wypiasnie i zapetlenie kazdego slowa z listy
  occurrences[word] = occurrences.get(word, 0) + 1 # sprawdzenie czy dany klucz istnieje, je≈ºeli nie to 0, jezeli tak to + 1

for word in occurrences:
  print('The word', word,'occurs',occurrences[word],'times in the string') # wypisanie dictionary
```
