structural pattern matching with match

```python
x = [123123]
match x:
  case 'A':
    print('A')
  case str():
    print('some other string')
  case 0:
    print('zero')
  case 1 | 2 | 3:
    print(f"value {x} in range")
  case int():
    print(f"the int is: {x}")
  case _:
    print('neither string or int')
```

```python
response = ''
while response != 'Q':
  response = input('Q to quit, b to break')
  if response == 'B':
    break
  else:
    print('cos tam cos tam')
```

the for loop

```python
x = [1.0,2.0,3.0,0.0,-1.0]
for n in x:
  if n == 0.0:
    print('skipping 0.0')
  elif n>0:
    print('no negative minuses')
    print(1/n)
  else:
    print('loo completed')
```

the range function

```python
x = [1,3,-7,4,9,-5,4]
for i in range(len(x)):
  if x[i] < 0:
    print("foud negative number at index",i)
```

controlling range with starting and stepping values

```python
x = list(range(3,7))
y = list(range(2,10))
z = list(range(5,3))
print(x,y,z)
x = list(range(0,10,2))
y = list(range(5,0,-1))
print(x,y)
```

```python
x = [1,2,3,4]
x_squared = (item * item for item in x) # taki zapis nie zajmuje duzo miejsca
print(x_squared)
for square in x_squared:
  print(square,)
```
