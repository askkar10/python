basic function definitions

```python
# def name(parameter1,parameter2,...):
#   body


def fact(n):
    '''Return the factorial of the given number.''' # opcjonalny opis
    r = 1
    while n > 0:
        r = r * n
        n = n - 1
    return r # wartosc zwrocona do funckji

print(fact(4)) # wypisanie wartosci funkcji
x = fact(4) # przypisanie wartosci funkcji do zmiennej x
print(x)
```

positional parameter

```python
def power(x,y):
  r=1
  while y>0:
    r = r*x
    y = y-1
  return r
print(power(3,3)) # jezeli wypiszemy tylko same 3 to wtedy bedzie blad, poniewaz musimy do funkcji podac 2 prametry

# default values
def power(x,y=2): # tutaj mamy podany default value dla y
  r=1
  while y>0:
    r=r*x
    y=y-1
  return r
print(power(4)) # wiec przy wypisanu funkcji mozemy uzyc tylko jednego parametru
print(power(y=5,x=2)) # jezeli podajemy default value to mozemy nawet zmienic kolejnosc
```

variable numbers of arguments

```python
# *args, **kwargs

def maximum(*numbers): # mozemy podac dowolna ilosc parametrow
  if len(numbers) == 0:
    return None
  else:
    maximum = numbers[0] # numbers to lista
    for n in numbers[1:]:
      if n > maximum:
        maximum =n
    return maximum

print(maximum(3,2,6,-1))
print(maximum(1,5,9,-2,2))

def example_fun(x,y,**other): # lapie wszystkie nazwane parametry
  print(f"x : {x}, y: ={y}, keys in 'other' : {list(other.keys())}")
  other_total = 0
  for k in other.keys(): # te parametry sa jako dictionary
    other_total = other_total + other[k]
  print(f"The total of values in 'other is {other_total}")

example_fun(2,y='1',foo=3,bar=4,ajj=10)
```

local, nonlocal and global variables

```python
def fun():
  global a
  a = 1 # jest to teraz globalna zmienna
  b = 2
fun()
print(a)

def myfunc():
  x = 'john'
  def myfunc2():
    nonlocal x
    x = 'hello'
  myfunc2()
  return x

print(myfunc())
```

```python
g_var = 0               #A
nl_var = 0               #B
print("top level-> g_var: {0} nl_var: {1}".format(g_var, nl_var))
def test():
    nl_var = 2              #C
    print("in test-> g_var: {0} nl_var: {1}".format(g_var, nl_var))
    def inner_test():

        global g_var         #D
        nonlocal nl_var      #E
        g_var = 1
        nl_var = 4
        print("in inner_test-> g_var: {0} nl_var: {1}".format(g_var,
                                                              nl_var))

    inner_test()
    print("in test-> g_var: {0} nl_var: {1}".format(g_var, nl_var))

test()
print("top level-> g_var: {0} nl_var: {1}".format(g_var, nl_var))
```

assignment functions to variables

```python
def f_to_kelvin(degrees_f):
  return 273.15 + (degrees_f - 32) * 5 / 9

def c_to_kelvin(degrees_c):
  return 273.15 + degrees_c

abs_temperature = f_to_kelvin # przypisanie funkcji do zmiennej
print(abs_temperature(32))
abs_temperature = c_to_kelvin # pzrypisanie funkcji do zmiennej
print(abs_temperature(0))

# funkcje w dictionary
t = {'FtoK': f_to_kelvin,'CtoK': c_to_kelvin}
print(t['FtoK'](32))
print(t['CtoK'](0))
```

generator functions

```python
# yeild
def four():
  x=0
  while x<=4:
    print(f'in generator, x={x}')
    yield x
    x +=1

for i in four():
  print(f'value from genetaror {i}')

# yield from
def subgen(x):
    for i in range(x):
        yield i

def gen(y):
    yield from subgen(y)

for q in gen(6):
    print(q)

2 in four()
5 in four()
```

decorators

```python
def decorate(func):
  print('in decorate function, decorating',
        func.__name__)
  def wrapper_func(*args):
    print('executin',func.__name__)
    return func(*args)
  return wrapper_func

@decorate
def myfunction(parameter):
  print(parameter)
myfunction('hello')

# A simple decorator function
def decorator(func):

    def wrapper():
        print("Before calling the function.")
        func()
        print("After calling the function.")
    return wrapper

# Applying the decorator to a function
@decorator

def greet():
    print("Hello, World!")

greet()
```
